```python
import random
import csv

# Define feature values
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
times = ["breakfast", "lunch", "dinner"]
vegetables = ["ghobi", "paneer makhni", "aloo sabji", "mix veg", "mushroom mattar", "methi malai", "bhindi"]
lentils = ["arhar", "moong", "chane", "masoor", "tur", "rajma", "makhni"]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Define conditions for normal day final demand:
# High-demand for paneer makhni or if lentils are makhni/rajma
# Lower-demand for aloo sabji or ghobi or if lentils are arhar/moong
rows = []
rows.append(["month", "time", "Vegetable", "Lentils", "day", "event", "holiday", "finaldemand"])

for _ in range(1000):
    month = random.choice(months)
    time_val = random.choice(times)
    veg = random.choice(vegetables)
    lentil = random.choice(lentils)
    day = random.choice(days)

    # Event: ~10% chance to be "yes"
    event = "yes" if random.random() < 0.10 else "no"

    # Holiday: "yes" for June and July (summer vacations)
    holiday = "yes" if month in ["June", "July"] else "no"

    # If it's a holiday or an event, ensure final demand is not more than 50
    if holiday == "yes" or event == "yes":
        finaldemand = random.randint(45, 50)
    else:
        # On normal days, set demand based on the dish type
        if veg == "paneer makhni" or lentil in {"makhni", "rajma"}:
            finaldemand = random.randint(195, 205)
        elif veg in {"aloo sabji", "ghobi"} or lentil in {"arhar", "moong"}:
            finaldemand = random.randint(145, 155)
        else:
            finaldemand = random.randint(190, 210)

    rows.append([month, time_val, veg, lentil, day, event, holiday, finaldemand])

# Write the data to a CSV file
with open("mess_dataset.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rows)

print("Dataset 'mess_dataset.csv' generated successfully.")
```

```python
# ================
# DATA GENERATION
# ================
import random
import csv

# Define feature values
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
times = ["breakfast", "lunch", "dinner"]
vegetables = ["ghobi", "paneer makhni", "aloo sabji", "mix veg",
             "mushroom mattar", "methi malai", "bhindi"]
lentils = ["arhar", "moong", "chane", "masoor", "tur", "rajma", "makhni"]
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

# Generate dataset
rows = []
rows.append(["month", "time", "Vegetable", "Lentils", "day", "event", "holiday", "finaldemand"])

for _ in range(1000):
    month = random.choice(months)
    time_val = random.choice(times)
    veg = random.choice(vegetables)
    lentil = random.choice(lentils)
    day = random.choice(days)

    event = "yes" if random.random() < 0.10 else "no"
    holiday = "yes" if month in ["June", "July"] else "no"

    if holiday == "yes" or event == "yes":
        finaldemand = random.randint(45, 50)
    else:
        if veg == "paneer makhni" or lentil in {"makhni", "rajma"}:
            finaldemand = random.randint(195, 205)
        elif veg in {"aloo sabji", "ghobi"} or lentil in {"arhar", "moong"}:
            finaldemand = random.randint(145, 155)
        else:
            finaldemand = random.randint(190, 210)

    rows.append([month, time_val, veg, lentil, day, event, holiday, finaldemand])

with open("mess_dataset.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(rows)

# ================
# MODEL TRAINING
# ================
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Feature engineering function
def enhance_features(df):
    # Create binary flags for special dishes
    df['high_demand_dish'] = ((df['Vegetable'] == 'paneer makhni') |
                             (df['Lentils'] == 'makhni') |
                             (df['Lentils'] == 'rajma')).astype(int)

    df['low_demand_dish'] = ((df['Vegetable'] == 'aloo sabji') |
                            (df['Vegetable'] == 'ghobi') |
                            (df['Lentils'] == 'arhar') |
                            (df['Lentils'] == 'moong')).astype(int)

    # Create interaction features
    df['holiday_or_event'] = ((df['holiday'] == 'yes') |
                             (df['event'] == 'yes')).astype(int)
    return df

# Load and preprocess data
df = pd.read_csv("mess_dataset.csv")
df = enhance_features(df)

# Prepare features and target
X = df.drop(columns=["finaldemand"])
y = df["finaldemand"]

# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

# Feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train model
rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=15,
    min_samples_leaf=4,
    random_state=42
)
rf.fit(X_train_scaled, y_train)

# ======================
# CUSTOM PREDICTION LOGIC
# ======================
def custom_predict(model, X_scaled, X_unscaled):
    preds = model.predict(X_scaled)

    for i in range(len(preds)):
        # Check holiday/event status
        is_holiday = 'holiday_yes' in X_unscaled.columns and X_unscaled.iloc[i]['holiday_yes'] == 1
        is_event = 'event_yes' in X_unscaled.columns and X_unscaled.iloc[i]['event_yes'] == 1

        if is_holiday or is_event:
            # Enforce hard cap for holidays/events
            preds[i] = min(preds[i], 50)
        else:
            # Apply boosts for special dishes
            boost = 0
            if 'Vegetable_paneer makhni' in X_unscaled.columns and X_unscaled.iloc[i]['Vegetable_paneer makhni'] == 1:
                boost += 20
            if 'Lentils_makhni' in X_unscaled.columns and X_unscaled.iloc[i]['Lentils_makhni'] == 1:
                boost += 15
            if 'Lentils_rajma' in X_unscaled.columns and X_unscaled.iloc[i]['Lentils_rajma'] == 1:
                boost += 15
            preds[i] += boost

            # Enforce reasonable ranges
            if (X_unscaled.iloc[i]['high_demand_dish'] == 1) and (preds[i] < 190):
                preds[i] = 195
            elif (X_unscaled.iloc[i]['low_demand_dish'] == 1) and (preds[i] > 160):
                preds[i] = 150

    return preds

# ================
# MODEL EVALUATION
# ================
y_pred = custom_predict(rf, X_test_scaled, X_test)

print("\n=== Model Evaluation ===")
print(f"MAE: {mean_absolute_error(y_test, y_pred):.2f}")
print(f"RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}")

# ====================
# SAMPLE PREDICTION
# ====================
def prepare_sample(sample):
    sample_df = pd.DataFrame([sample])
    sample_df = enhance_features(sample_df)
    sample_encoded = pd.get_dummies(sample_df, drop_first=True)

    # Ensure all columns match training data
    for col in X_encoded.columns:
        if col not in sample_encoded:
            sample_encoded[col] = 0

    # Reorder columns to match training data
    sample_encoded = sample_encoded[X_encoded.columns]

    # Manual override for holiday/event columns
    sample_encoded['holiday_yes'] = 1 if sample['holiday'] == 'yes' else 0
    sample_encoded['event_yes'] = 1 if sample['event'] == 'yes' else 0

    return sample_encoded

# Test case 1: Normal day with popular dishes
sample_input = {
    'month': 'April',
    'time': 'dinner',
    'Vegetable': 'paneer makhni',
    'Lentils': 'rajma',
    'day': 'Wednesday',
    'event': 'no',
    'holiday': 'no'
}

# Test case 2: Holiday
sample_input2 = {
    'month': 'June',
    'time': 'lunch',
    'Vegetable': 'aloo sabji',
    'Lentils': 'moong',
    'day': 'Saturday',
    'event': 'no',
    'holiday': 'no'
}

for i, sample in enumerate([sample_input, sample_input2], 1):
    prepared_sample = prepare_sample(sample)
    scaled_sample = scaler.transform(prepared_sample)
    prediction = custom_predict(rf, scaled_sample, prepared_sample)

    print(f"\nSample {i} Prediction:")
    print("Input Features:")
    for k,v in sample.items():
        print(f"- {k}: {v}")
    print(f"Predicted Demand: {prediction[0]:.1f}")
```
